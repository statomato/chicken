```python
import numpy as np
def round_005(num):
    num_01 = np.round(num, 2)
    num_range = np.array([num_01-0.005, num_01, num_01+0.005])
    
    num_dist = (num - num_range)**2
    
    num_min = np.argmin(num_dist)
    num_out = num_range[num_min]
    
    return num_out
```

```python
chicken['lat'] = chicken['위도'].apply(round_005)
chicken['lon'] = chicken['경도'].apply(round_005)
```
